```python
p_array_copy = deepcopy(parray)
nparts_in_slice=5000
smooth_param=0.01
nbins_x=200
nbins_y=200
interpolation="bilinear"
inverse_tau=False,
filter_base=2
filter_iter=2
figname=40,
slice_params = global_slice_analysis(p_array_copy, nparts_in_slice, smooth_param, filter_base, filter_iter)
#plt.title("Energy spread")
plt.plot(slice_params.s*1e3, slice_params.I, "b")
plt.plot(slice_params.s*1e3,[4000]*len(slice_params.s))
# plt.legend()
plt.xlabel(r"$s\,[\mathrm{mm}]$")
plt.ylabel(r"$I\,[\mathrm{A}]$")
plt.xlim((0.025,0.065))
plt.show()
```
